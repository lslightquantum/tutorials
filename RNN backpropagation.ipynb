## RNN Model
$$
\mathbf{h}^{(t)}=tanh(\mathbf{b}+\mathbf{W_h}\mathbf{h}^{(t-1)}+\mathbf{W_x}\mathbf{x}^{(t)})
$$
$$
\mathbf{o}^{(t)}=\mathbf{c}+\mathbf{W_y}\mathbf{h}^{(t)}
$$
$$
\mathbf{\hat{y}}^{(t)}=softmax(\mathbf{o}^{(t)})
$$
## Loss Function
Loss at time t is
$$
\mathbf{L}^{(t)}=-\sum_i
y^{(t)}_i\log \hat{y}^{(t)}_i
$$

## Partial Derivative w.r.t. $\mathbf{W_y}$ and $\mathbf{c}$
$$
\frac{\partial L^{(t)}}{\partial \mathbf{W_y}}=
\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{W_y}}
$$
$$
\frac{\partial L^{(t)}}{\partial \mathbf{c}}=
\frac{\partial L^{(t)}}{\partial \mathbf{\mathbf{o}^{(t)}}}
\frac{\partial \mathbf{\mathbf{o}^{(t)}}}{\partial \mathbf{c}}=
\frac{\partial L^{(t)}}{\partial \mathbf{\mathbf{o}^{(t)}}}
$$
Now, calculate partial derivative w.r.t. $\mathbf{o}^{(t)}$.
$$
\frac{\partial \hat{y}^{(t)}_j}{\partial o^{(t)}_i}=
\left\{
\begin{array}{rcl}
\hat{y}^{(t)}_i(1-\hat{y}^{(t)}_i) && {i=j}\\
-\hat{y}^{(t)}_i\hat{y}^{(t)}_j  && {i\neq j}\\
\end{array} \right.
$$
$$
\begin{align}
\frac{\partial L^{(t)}}{\partial o^{(t)}_i}&=-\sum_k
\frac{y^{(t)}_k}{\hat{y}^{(t)}_k}
\frac{\partial \hat{y}^{(t)}_k}{\partial o^{(t)}_i}\\
&=-y^{(t)}_i(1-\hat{y}^{(t)}_i)-\sum_{k\neq i}\frac{y^{(t)}_k}{\hat{y}^{(t)}_k}(-\hat{y}^{(t)}_k\hat{y}^{(t)}_i)\\
&=-y^{(t)}_i+y^{(t)}_i\hat{y}^{(t)}_i+\sum_{k\neq i}{y^{(t)}_k}\hat{y}^{(t)}_i\\
&=-y^{(t)}_i+\hat{y}^{(t)}_i\sum_{k}{y^{(t)}_k}\\
&=\hat{y}^{(t)}_i-y^{(t)}_i
\end{align}
$$
Therefore,
$$
\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}=
\mathbf{\hat{y}}^{(t)}-\mathbf{y}^{(t)}
$$
Finally,
$$
\frac{\partial L^{(t)}}{\partial \mathbf{W_y}}=
(\mathbf{\hat{y}}^{(t)}-\mathbf{y}^{(t)}){h^{(t)}}^T
$$
and
$$
\frac{\partial L^{(t)}}{\partial \mathbf{c}}=
\mathbf{\hat{y}}^{(t)}-\mathbf{y}^{(t)}
$$
Derivatives w.r.t. total loss are
$$
\frac{\partial L}{\partial \mathbf{W_y}}=\sum_t
(\mathbf{\hat{y}}^{(t)}-\mathbf{y}^{(t)}){h^{(t)}}^T
$$
$$
\frac{\partial L}{\partial \mathbf{c}}=\sum_t
(\mathbf{\hat{y}}^{(t)}-\mathbf{y}^{(t)})
$$

## Partial Derivative w.r.t. $\mathbf{W_h}$, $\mathbf{W_x}$, and $\mathbf{b}$
$$
\frac{\partial L^{(t)}}{\partial \mathbf{W_h}}
=\sum\limits_{k=0}^{t-1}\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{h}^{(t)}}
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}
\frac{\partial \mathbf{h}^{(k)}}{\partial \mathbf{W_h}}
$$
$$
\frac{\partial L^{(t)}}{\partial \mathbf{W_x}}
=\sum\limits_{k=0}^{t-1}\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{h}^{(t)}}
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}
\frac{\partial \mathbf{h}^{(k)}}{\partial \mathbf{W_x}}
$$
$$
\frac{\partial L^{(t)}}{\partial \mathbf{b}}
=\sum\limits_{k=0}^{t-1}\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{h}^{(t)}}
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}
\frac{\partial \mathbf{h}^{(k)}}{\partial \mathbf{b}}
$$
where
$$
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}=
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(t-1)}}...
\frac{\partial \mathbf{h}^{(k+1)}}{\partial \mathbf{h}^{(k)}}=
\prod\limits_{i=k}^{t-1}\frac{\partial \mathbf{h}^{(i+1)}}{\partial \mathbf{h}^{(i)}}
$$
and,
$$
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(t-1)}}=
\mathbf{W}^T_\mathbf{h}\mathrm{diag}(\mathbf{1}-{\mathbf{h}^{(t)}}^2)
$$
Derivatives w.r.t. total loss are
$$
\begin{align}
\frac{\partial L}{\partial \mathbf{W_h}}&=\sum_t
\sum\limits_{k=0}^{t-1}\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{h}^{(t)}}
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}
\frac{\partial \mathbf{h}^{(k)}}{\partial \mathbf{W_h}}
\\
\frac{\partial L}{\partial \mathbf{W_x}}&=\sum_t
\sum\limits_{k=0}^{t-1}\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{h}^{(t)}}
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}
\frac{\partial \mathbf{h}^{(k)}}{\partial \mathbf{W_x}}
\\
\frac{\partial L}{\partial \mathbf{b}}&=\sum_t
\sum\limits_{k=0}^{t-1}\frac{\partial L^{(t)}}{\partial \mathbf{o}^{(t)}}
\frac{\partial \mathbf{o}^{(t)}}{\partial \mathbf{h}^{(t)}}
\frac{\partial \mathbf{h}^{(t)}}{\partial \mathbf{h}^{(k)}}
\frac{\partial \mathbf{h}^{(k)}}{\partial \mathbf{b}}
\end{align}
$$
